```python
import pandas as pd
cities = pd.read_csv('../datasets/python_data_science_handbook/california_cities.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lat, lon = cities['latd'], cities['longd']
```

```python
population, area = cities['population_total'], cities['area_total_km2']
```

```python
plt.scatter(lon, lat, label=None,
           c = np.log10(population), cmap='viridis',
           s=area, linewidth=0, alpha=0.5)

plt.axis(aspect = 'equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)

# insert labels for area, and set its size = area
# remaining points keep the same unit and no changes
# label value is relative to the values of the plots
for area in [100,300,500,800]:
    plt.scatter([],[],c='k', alpha=0.3, s=area,
               label=str(area) + 'km$^2$')
    
plt.legend(scatterpoints=1, frameon=False,
          labelspacing=1, title='City Area')

plt.title('California Cities: Area and Population');
```
