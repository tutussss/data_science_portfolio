##### we want to see if there is any linear dependancy 

##### X is independant variable  
take our independant variables which are in first 4 columns

```python
# iloc[].values returns an array matrix
X = dataset.iloc[:,:-1].values
```

##### Y is dependant variable we are trying to predict 

```python
y = dataset.iloc[:,4].values
```

```python
y
```

### we will have to transform the categorical variables 

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
labelencoder_X = LabelEncoder()
```

##### One hot encoder needs number for encoding but we have string, so we first encode to integer then we do One Hot Encoding

```python
X[:,3] = labelencoder_X.fit_transform(X[:,3])
```

```python
onehotencoder = OneHotEncoder(categorical_features=[3])
```

```python
X = onehotencoder.fit_transform(X).toarray()
```

```python
X
```

#### avoiding dummy variable trap for libraries that doesn't care about dummy variable
X= X[:,1:]

### Split into train and test data 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_train.shape
```

```python
y_train
```

## TRAIN THE MODEL 

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

## BACKWARD ELIMINATION 

```python
import statsmodels.api as sm
```

```python
X = np.append(arr = np.ones((50,1)).astype(int), values = X, axis =1)
```

```python
X
```

```python
# X_opt with all the 6 features
X_opt = X[:,[0,1,2,3,4,5]]
```

##### SL = 0.05 

##### new regressor but from Stats Model 

```python
regressor_OLS = sm.OLS(endog=y, exog=X_opt).fit()
```

we fitted full model with all possible predictors

now we check p values 

```python
regressor_OLS.summary()
```
