```python
import time 

a = np.random.rand(10000000)
b = np.random.rand(10000000)

tlc = time.time()
c = np.dot(a,b)
toc = time.time()

print(str((toc-tlc)*100) + 'ms')
```

```python
c = 0 
tlc1 = time.time()
for i in range(10000000):
    c += a[i]*b[i]
toc1= time.time()


print(str((toc1-tlc1)*100) + 'ms')
```

```python
print(str((toc1-tlc1)/(toc-tlc)) + ' times faster')
```
