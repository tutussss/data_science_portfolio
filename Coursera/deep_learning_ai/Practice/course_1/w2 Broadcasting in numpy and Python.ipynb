```python
import numpy as np

A = np.array([[56.0, 0.0, 4.4, 68.0],
              [1.2, 104.0, 0.52, 8.0],
              [1.8, 135.0, 0.99, 0.9]])

print(A.shape)
```

sum vertically using axis=0, axis = 0 is vertically <br>
sum horizontally using axis=0, axis = 0 is horizontally

```python
cal = A.sum(axis = 0)
print(cal)
```

```python
cal.shape
```

```python
# reshape is just to make sure otherwise the array was already in shape
percentage = 100*A/cal.reshape(1,4)
print(percentage)
```

##### MORE 

broadcasting and dot product is affected by shape of the vectors, keep and eye out on it

```python
a = np.random.randn(5)
```

a = np.random.randn(5) creates a rank 1 array which has shape (5,)

```python
print(a)
```

```python
print(a.shape)
```

```python
print(a.T)
```

```python
print(np.dot(a,a.T))
```

```python
a = np.random.randn(5,1) # take vector of shape 5,1
print(a)
```

```python
print(a.T)
```

```python
print(np.dot(a,a.T))
```

#### we can also user <b>'assert(a.shape==(5,1))'</b> to make sure that the array is of certain shape <br>
#### we can also reshape the rank 1 array<br>
#### avoid rank 1 array always
