to visualize the corelation, use Seaborn.regplot()

```python
sb.regplot(data=fuel_econ, x='displ', y='comb')
```

we can remove regression line using fit_reg=False

```python
sb.regplot(data=fuel_econ, x='displ', y='comb', fit_reg=False)
```

the relation seems to be logarithmic 

```python
def log_transform(x, inverse=False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10,x)    
```

```python
sb.regplot(x=fuel_econ.displ, y=fuel_econ.comb.apply(log_transform))
tick_locs= [10,20,50,100,200,500]
plt.yticks(log_transform(tick_locs), tick_locs)
```

##### we need to do some summarization before plotting heatmap for categorical variables 

merging different transmission types into two categories Auto and manual

```python
fuel_econ_2trans=fuel_econ
fuel_econ_2trans['trans']=np.where(fuel_econ.trans.str.match('Auto*'),'Automatic','Manual')
fuel_econ_2trans
```

```python
ct_counts = fuel_econ_2trans.groupby(['VClass','trans']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts = ct_counts.pivot(index='VClass',
                            columns='trans', values='count')
```

```python
sb.heatmap(ct_counts)
```

<b>fmt</b> and <b>annot</b> properties makes the heatmap to show the values inside

```python
sb.heatmap(ct_counts, annot=True, fmt='d',)
```

## ADAPTED BAR CHARTS 

```python
base_color=sb.color_palette()[0]
sb.barplot(data=fuel_econ, x='VClass', y='comb',
              color=base_color)
plt.xticks(rotation = 15)
plt.ylabel('Avg combines fuel eff')
```

remove error bars by errwirdth=0

```python
base_color=sb.color_palette()[0]
sb.barplot(data=fuel_econ, x='VClass', y='comb', errwidth=0,
              color=base_color)
plt.xticks(rotation = 15)
plt.ylabel('Avg combines fuel eff')
```

to se standard deviation

```python
base_color=sb.color_palette()[0]
sb.barplot(data=fuel_econ, x='VClass', y='comb', errwidth=0, ci='sd' ,
              color=base_color)
plt.xticks(rotation = 15)
plt.ylabel('Avg combines fuel eff')
```

if lower portion isn't of interest, remove it by pointplot

```python
base_color=sb.color_palette()[0]
sb.pointplot(data=fuel_econ, x='VClass', y='comb', 
             ci='sd' ,
              color=base_color)
plt.xticks(rotation = 15)
plt.ylabel('Avg combines fuel eff')
```

to remove the connected lines, add linestyles=""

```python
base_color=sb.color_palette()[0]
sb.pointplot(data=fuel_econ, x='VClass', y='comb', 
             ci='sd' ,
              color=base_color,
            linestyles="")
plt.xticks(rotation = 15)
plt.ylabel('Avg combines fuel eff')
```

##### additional plots practice: plot the average combined fuel efficiency for each manufacturer with at least 80 cars.

```python

    # data setup
    fuel_econ = pd.read_csv('./data/fuel_econ.csv')

    most_makes = fuel_econ['make'].value_counts().index[:18]
    fuel_econ_sub = fuel_econ.loc[fuel_econ['make'].isin(most_makes)]

    make_means = fuel_econ_sub.groupby('make').mean()
    comb_order = make_means.sort_values('comb', ascending = False).index

    # plotting
    g = sb.FacetGrid(data = fuel_econ_sub, col = 'make', col_wrap = 6, size = 2,
                     col_order = comb_order)
    # try sb.distplot instead of plt.hist to see the plot in terms of density!
    g.map(plt.hist, 'comb', bins = np.arange(12, fuel_econ_sub['comb'].max()+2, 2))
    g.set_titles('{col_name}')
```
