```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../datasets/Python-for-Finance-Repo-master/walmart_stock.csv', parse_dates=True, index_col='Date')
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

#####  Rolling takes average from previous data

```python
df.rolling(7).mean().head(14)
```

```python
df['Open'].plot()
df.rolling(window=7).mean()['Close'].plot(figsize=(16,6))
```

####  this techinique can be used to see the variations trends

```python
df['Open'].plot()
df.rolling(window=30).mean()['Close'].plot(figsize=(16,6))
```

### Trick for adding default legends, use df['new col'] to store previous plot

```python
df['Close 30 days ma'] = df['Close'].rolling(window=30).mean()
df[['Close 30 days ma', 'Close']].plot(figsize=(16,6))
```

## EXPANDING 
#### everything that came before, may not be useful for time series

```python
df['Close'].expanding().mean().plot(figsize=(16,6))
```

### BOLLINGER BAND 
##### volatility increase- 

```python
df['Close 20 day mean'] = df['Close'].rolling(20).mean()
df['Upper'] = df['Close 20 day mean'] + 2*(df['Close'].rolling(20).std())
df['Lower'] = df['Close 20 day mean'] - 2*(df['Close'].rolling(20).std())
df[['Close', 'Close 20 day mean', 'Upper', 'Lower']].plot(figsize=(20,10))
```
