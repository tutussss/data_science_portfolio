##### resample based on year, rule='A' 

```python
df.resample(rule='A').mean()
```

```python
# Business Quarter
df.resample(rule='BQ').mean()
```

#### We can provide ouw own function for resampling aggregation function

```python
def first_date(entry):
    return entry[0]
```

```python
df.resample('A').apply(first_date)
```

### resampling helps in different visualization 

```python
# 'A' is by year
df['Close'].resample('A').mean().plot(kind='bar')
```

```python
# 'M' is by month, not continuous plot but bar
df['Close'].resample('M').mean().plot(kind='bar', figsize=(16,6))
```
