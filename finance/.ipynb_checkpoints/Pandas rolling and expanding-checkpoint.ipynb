```python
df = pd.read_csv('../datasets/Python-for-Finance-Repo-master/walmart_stock.csv', parse_dates=True, index_col='Date')
```

```python
df.rolling(7).mean().head(14)
```

```python
df['Open'].plot()
df.rolling(window=7).mean()['Close'].plot(figsize=(16,6))
```
